```python
import os
import subprocess

# 创建目标文件夹（如果不存在）
os.makedirs("dataset", exist_ok=True)

# 文件链接列表
file_urls = [
    "https://snap.stanford.edu/data/cit-Patents.txt.gz",
    "https://snap.stanford.edu/data/email-Enron.txt.gz",
    "https://snap.stanford.edu/data/amazon0601.txt.gz",
    "https://snap.stanford.edu/data/roadNet-PA.txt.gz",
    "https://snap.stanford.edu/data/web-Google.txt.gz",
    "https://snap.stanford.edu/data/email-Eu-core.txt.gz"
]

# 下载文件
def download_files():
    for url in file_urls:
        print(f"Downloading {url}...")
        subprocess.run(["wget", "-P", "dataset", url])

# 解压缩文件
def extract_files():
    for file in os.listdir("dataset"):
        if file.endswith(".gz"):
            file_path = os.path.join("dataset", file)
            print(f"Extracting {file_path}...")
            subprocess.run(["gunzip", "-k", file_path])




download_files()
extract_files()


```
